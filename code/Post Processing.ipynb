```python
#import required libraries

import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
# read csv file for geometric features

result = pd.read_csv('result.csv',index_col=0)
result = result.T
result.columns = "AR WR PtoA PHtP AHTAL".split()
result.head()
```

```python
#read csv file for colour histogram

hist = pd.read_csv('histo.csv',index_col=0)
```

```python
#rename hist dataframe's columns

column = []
for i in range(0,768):
    column.append('c'+str(i))
    
hist.columns=column
```

```python
hist.head()
```

```python
#read csv file of Histogram of Gradients [HOG]

hog = pd.read_csv('hog_gray.csv',index_col=0)
cols = []
for i in range(len(hog.columns)):
    cols.append('hog'+str(i))
hog.columns=cols
```

```python
#add geometric features and hog to hist dataframe

hist['AR']=np.array(result['AR'])
hist['WR']=np.array(result['WR'])
hist['PtoA']=np.array(result['PtoA'])
hist['PHtP']=np.array(result['PHtP'])
hist['AHTAL']=np.array(result['AHTAL'])
for i in hog.columns:
    hist[i] = hog[i]
```

```python
hist
```

```python
#take all the indexes in the hist dataframe in a list index

index=hist.index
```

```python
#define a new dataframe to work upon

df=hist
```

```python
#write the final dataframe to final_result csv file

hist.to_csv('final_result.csv')
```

```python
#split the original dataframe into test and train dataframes

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df,index,
                                                    test_size=0.25,random_state=401)
```

```python
#import kNN classifier

from sklearn.neighbors import KNeighborsClassifier
```

```python
#calculate error rate of classification against the value of k

error_rate = []

# Will take some time
for i in range(1,20):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
#plot error rate vs k

plt.figure(figsize=(10,6))
plt.plot(range(1,20),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.savefig('error_vs_k.png')
```

```python
#choose k with minimum error

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
#train(fit) kNN model with X_train and y_train
 
knn.fit(X_train,y_train)
```

```python
#predict X_test

pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#generate the confusion matrix and classification report

print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
#calculate error rate of RFC model against n_estimators

error = []
i = 25
while i<=500:
    rfc = RandomForestClassifier(n_estimators=i)
    rfc.fit(X_train,y_train)
    pred_i = rfc.predict(X_test)
    error.append(np.mean(pred_i != y_test))
    i+=25
```
